### Beam Properties

```python
from indeterminatebeam import Beam
beam = Beam(9, E=2000, I=10**6, A=3000)
```

### Supports


```python
from indeterminatebeam import Support
a = Support(0, (1,1,0)) #pin
b = Support(3.5, (0,1,0)) #roller
c = Support(6, (0,1,0)) #fixed
d = Support(9, (1,1,1))
beam.add_supports(a,b,c,d)
```

### Loads

```python
from indeterminatebeam import PointLoadV, PointTorque, DistributedLoadV
load_1 = PointLoadV(-1000,5)
load_2 = DistributedLoadV(-10000,(0,7))
load_3 = PointTorque(1.5*10**3, 8)
beam.add_loads(load_1,load_2,load_3)
```

```python
beam.analyse()
```

### Plotting

```python
fig_1 = beam.plot_beam_external()
fig_1.show()

fig_2 = beam.plot_beam_internal()
fig_2.show()
```
